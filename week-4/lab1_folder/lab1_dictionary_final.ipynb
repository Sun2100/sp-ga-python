```python
import csv

#a dictionary
            #TODO dictionary should look like this 
            #{account number: entry#: title, expense type, amount, date
            # account number: entry#: title, expense type, amount, date
            # account number: entry#: title, expense type, amount, date
            # account number: entry#: title, expense type, amount, date}

            # account number: 10, entry#: 0, title: t, expese type: income, amount: 10, date: 1-1-2002

fields = ['title', 'expense_type', 'amount', 'date']
entry_list = []
income_list = []
expense_list = []

def get_input():
    title = input("Enter the title: ")
    expense_type = input("Enter expense type: ")
    amount = input("enter amount:")
    date = input("enter the date: ")

    #seperate the amount into positive and negative
    if int(amount) > 0:
        income_list.append(amount)
    else:
        expense_list.append(amount)


    add_new_entry(title,expense_type,amount,date)

    option = input("Enter another entry? (y/n)")

    return option

def add_new_entry(titl,exp,amt,date):
    account = {}
     
    account["title"] = titl
    account["expense_type"] = exp  
    account["amount"] = amt
    account["date"] = date

    entry_list.append(account)
    
       
def write_to_file():
    with open("./budget_output_solo_dictionary.csv", "w" ) as file:
        writer = csv.DictWriter(file, fieldnames=fields)
        writer.writeheader() 
        writer.writerows(entry_list)

def in_it(): #check if file exist
    
    try:
       data_file = open("./budget_output_solo_dictionary.csv", "r") 
       reader = csv.DictReader(data_file) #load the file to the list if file exist
       for row in reader:
           entry_list.append(row) #grab the old and add to new entry
       data_file.close() 
    except:
        print("will create file")

def view_all_previous_entries():
    data_file = open("./budget_output_solo_dictionary.csv", "r") 
    reader = csv.DictReader(data_file) #load the file to the list if file exist
    for row in reader:
        print(row)

def display_total_balance(): #add all income and subtract expense
#have 2 lists income and expense 
#add total income and expense 
#sub to find the difference 
    total = 0
    average = 0
    index = 0

    with open("./budget_output_solo_dictionary.csv", "r") as file:

        rows = []
        reader = csv.reader(file, delimiter= ",")

        for line in reader:    
            rows.append(line)
        
        #print(rows)
        rows.remove(rows[0])
        print(rows[2][2])
        
        for row in rows:  
            target = row[2]
            total += float(target)
            index += 1

        average = total // index

        print(f"total account balance is ${total} and the average transaction size is ${average}")

def run_it():
    in_it()
    
    while True:
        choices = get_input()
        if choices == "n":
            break
    write_to_file()

command = ""

#run_it()
#view_all_previous_entries()

while command != "q":
    command = input(
        "Welcome to the Budget Tracker\n\nPress [a] to add a new entry.\nPress [b] to display total account balance.\nPress [c] to view all previous entries.\nPress [q] to quit.")

    if command == "a":
        run_it()
    elif command == "b":
        view_all_previous_entries()
    elif command == "c":
        display_total_balance()
```

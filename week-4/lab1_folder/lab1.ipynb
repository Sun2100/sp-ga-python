```python
import os
import csv

#with open("lines.txt") as my_file:
#    lines = my_file.readlines()
#    print(lines)

def add_new_entry(account):
    title = ""
    expense_type = ""
    amount = 0
    date = ""
    income_or_expense = 1
 
    with open("./budget_output.csv", "a") as file:

        #write header TODO

        writer = csv.writer(file)


        title = input("Please enter a title: ")

        while expense_type != "e" and expense_type != "i":
            expense_type = input("Is this an [E]xpense or [I]ncome? ").lower()

        expense_type = "Income" if expense_type == "i" else "Expense"
        income_or_expense = 1 if expense_type == "Income" else -1

        amount = int(input("enter amount: ")) * income_or_expense

        #if expense_type == "i":  
        #    amount = input("Please enter an amount: ")
        #else:
        #    amount = -1*(input("Please enter an amount: "))

        date = input("Please enter a date: ")

        #might need .strip() at the end except amount TODO

        writer.writerow([account, title, expense_type, amount, date]) #key line

        file.flush()

def display_total_account(account): #create a class with list that gets update alone with writing into the block

    total = 0
    account = int(account)
    average = 0
    index = 0

    with open("./budget_output.csv", "r") as file:

        rows = []
        reader = csv.reader(file, delimiter= ",")

        for line in reader:
            rows.append(line)

        for row in rows:
            if int(row[0]) == account:
                total += float(row[3])
                index += 1

        average = total // index

        print(f"total account balance for account number {account} is ${total} and the average transaction size is ${average}")


def display_all_account_entries():
    with open("./budget_output.csv", "r") as file:
        rows = []
        reader = csv.reader(file)

        for line in reader:
            rows.append(line)

        print("A summary of all accounts\n----------------------------------\n")

        for row in rows:
            print(f"title: {row[0]}\texpense type: {row[1]}\t amount: {row[1]}\t date: {row[2]}")

        print("\n")

command = ""
account = 0

while command != "q":
    command = input(
        "Welcome to the Budget Tracker\n\nPress [a] to add a new entry.\nPress [b] to display total account balance.\nPress [c] to view all previous entries.\nPress [q] to quit.")

    if command == "a":
        account = input("enter your account: ")
        if account != "q":
            add_new_entry(account)
        else:
            pass

    elif command == "b":
        account = input("enter your account: ").lower()
        if account != "q":
            display_total_account(account)
        else:
            pass

    elif command == "c":
        display_all_account_entries()
```

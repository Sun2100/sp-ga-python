```python
#create dataframe
data = {"Make": ["Toyota", "Lexus", "BMW", "Hyundai"], 
        "Model": ["Rav 4", "LS20", "X3", "Elantra"],
        "Year": [2018, 2015, 2012, 2022], 
        "Color": ["White", "Black", "Grey", "Blue"]}
```

```python
import pandas as pd

cars = pd.DataFrame(data)
```

```python
cars
```

```python
more_data = {"Make": ["Toyota", "Lexus", "BMW", "Hyundai"], 
        "Year": [2022, 2021, 2018, 2220],
        "Model": ["Corolla", "LS50", "X6", "Not Sure"], 
        "Color": ["White", "Black", "Grey", "Blue"]}

more_cars = pd.DataFrame(more_data)
```

```python
#combine both databases
#TODO CONCATINATIONS

pd.concat([cars, more_cars])
```

```python
price_data = {"Price": [35000, 50000, 54000, 27000], "Seats": [5, 5, 5, 5]}
prices = pd.DataFrame(price_data)
```

```python
pd.concat([cars, prices], axis=1) #axis = 1 means add more columns
#axis = 0 means adding rows
```

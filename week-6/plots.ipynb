```python
# Fixing random state for reproducibility
import numpy as np
np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

#.barh for horizontal bar chart
ax.barh(y_pos, performance, xerr=error, align='center')
#setting title [ticks] (tom, dick, harry, slim, jim) 
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()

```

```python
#artist record sales for the year
artists = ["Cardi B", "Drake", "The Stallion", "Lil Baby", "Camailla Cabello"]
sales = [150, 300, 650, 400, 950]

#plot these on a bar chart

#import matplotlib.pyplot as plt

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots() #subplots allows you to have multiple plots on the same plot 
#ex: can plot for male and female artist plot in one graph.

#ax.bar(labels, men_means, width, yerr=men_std, label='Men') yerr is the standard deviation
ax.bar(artists, sales, width, label='Artists')

ax.set_ylabel('Sales')
ax.set_title('Sales by popular artists')
ax.legend()

plt.show()


#ANS:
#plot these on a bar chart
#fig, g = plt.subplots()
#g.bar(artists, sales, 0.3 (this is width), label="Artist")

#lables
#ax.set_ylabel('Sales')
#ax.set_title('Sales by popular artists')

#plt.show




```
